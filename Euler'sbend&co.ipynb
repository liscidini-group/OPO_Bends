# Bends for FLM

## A little bit of math

#### Euler's bend

Determining curvature or curvature radius of a bend is the key to determining the bend geometry. For the Euler bend, the relationship between the bend curvature K (or the curvature radius ρ) at one bend point and bend path length s is expressed as follows:

\begin{equation}
K(s)=\dfrac{1}{\rho}=ks
\end{equation}

where k is the scaling factor. By introducing an adjustable parameter n, the NA (n-adjustable) bend is developed. The relationship between K and s for this bend is expressed as follows:

\begin{equation}
K(s)=\dfrac{1}{\rho}=ks^n
\end{equation}

It can be seen form (2) that the curvature K varies exponentially with the path length s. By varying the values of bend parameter n, a series of NA bends with different geometries can be obtained.
The corresponding differential expression of the curvature can be expressed as follows, according to the definition of curvature,
\begin{equation}
K(s)=\dfrac{d\theta(s,n)}{ds}
\end{equation}
where s is the path length along the bend and θ (s, n) is the angle of the tangent direction.

As shown in the figure, for one bend point P, the path length from the starting point to P is $s_p$; then, the angle of the tangent direction at P can be calculated by integration of K(s)

\begin{equation}
\theta(s_p,n)=\int_0^{s_p}K(s)ds = \int_0^{s_p} ks^n ds = \dfrac{k s^{n+1}_p}{n+1}
\end{equation}

Finally, every point $(x,y)$ on the curve is described by the Fresnel integral

\begin{equation*}
x(s_p)  = \int_0^{s_p} \cos(\theta(s,n)) ds = \int_0^{s_p} \cos \left(\dfrac{ks^{n+1}}{n+1}\right) ds
\end{equation*}

\begin{equation*}
y(s_p)  = \int_0^{s_p} \sin(\theta(s,n)) ds = \int_0^{s_p} \sin \left(\dfrac{ks^{n+1}}{n+1}\right) ds
\end{equation*}

that can be solved by numerical integration. For a determined effective bending radius R, an optimal NA bend with ultra-low bend loss can be obtained by scanning the bend parameter n.

Let us now consider two different cases: in the first one we will calculate and plot the curves without any particular costraints. In the second case we will focus on 90° bends, that can be obtained by fixing the initial and final point of the curve and defining an effective radius $R_{eff}$

#### N-Adjustable Euler's bend

Let us now calculate the constant k by considering the fact that one can write

\begin{equation*}
\dfrac{1}{R}=\,k L^n
\end{equation*}

where $R$ and $L$ are the final radius and final lenght of the curve. Knowing that $K=\dfrac{d\theta}{d s}$, we can obtain that

\begin{equation*}
\dfrac{d\theta}{ds}=\dfrac{s^n}{R L^n} 
\end{equation*}

which lead us to
\begin{equation*}
\theta=\dfrac{L^{n+1}}{R L^n (n+1)}=\dfrac{L}{R(n+1)}
\end{equation*}

and the final length of the curve can be written as
\begin{equation*}
L=(n+1)\theta R
\end{equation*}


### Symmetric 90° Euler's bend

For the 90° bends, we want to obtain the expression of the curve for a path that goes from (0,0) to ($R_{eff}.R_{eff}$). The shortest path will be the circular one (n=0), while the Euler and the N-adjustable Euler will change their path accordingly to the constant n and k. Analougsly, we can also control the Bezier by the parameter B. 

Two 45° NA bends can generate a 90° NA bend by splicing. For an 45° NA bend, when $\theta=0$ equals $\pi/4$, we can deduced that
\begin{equation*}
\theta(s_{sp},n)=\int_o^{s_p} k s^n ds = \dfrac{k s_p^{n+1}}{n+1}=\dfrac{\pi}{4}
\end{equation*}
We can derive the path length $s_p$ of the 45° NA bend
\begin{equation*}
s_p(k,n)=\left( \dfrac{\pi(n+1)}{4k} \right)^{\dfrac{1}{n+1}}
\end{equation*}
For the bend endpoint, the relationship between the coordinates $x(s_p)$, $y(s_p)$ and the radius $R_{eff}$ is
\begin{equation*}
x_{\theta=\pi/4}+y_{\theta=\pi/4}=R
\end{equation*}
which becomes
\begin{equation*}
\int_0^{s_p(k,n)} \cos\left(\dfrac{k s^{n+1}}{n+1}\right)ds + \int_0^{s_p(k,n)}\sin\left(\dfrac{k s^{n+1}}{n+1}\right)ds=R
\end{equation*}
Then substituting the expression for $s_p$, we get a condition to determine the parameter k. 
The analytical solution of this equation is almost non-existent, and the k value can be determined by performing trial calculations (constantly changing the k value for trial calculations until the condition is satisfied). 
It is worth pointing out that in previous study on Euler bend, the scaling factor k was set to 1.0 and bends corresponding to different effective bending radius were generated by introducing an index A. 

\begin{equation*}
\int_0^{s_p(k=1,n)} A \cos\left(\dfrac{ s^{n+1}}{n+1}\right)ds + \int_0^{s_p(k=1,n)}A \sin\left(\dfrac{s^{n+1}}{n+1}\right)ds=R
\end{equation*}

This description is not rigorous enough, although the obtained bends are correct.

### Hybrid Euler's bend

To connect two straight waveguides separated by distance d,a. 180∘ bend consisting of two Euler bend sections and a circular bend section can be used. The two Euler bend sections connect each straight waveguide to the circular bend of angle θ ∈ (0, π), allowing for adiabatic mode conversion from infinite radius (R = ∞) to a finite value 
\begin{equation*}
R_F=\dfrac{1}{\sqrt{k\pi p}}
\end{equation*}

Here, the Euler portion factor p, denoting the ratio of two Euler bends in the total $\pi$-bend, is defined as $p=1-\theta/pi$. We emphasize that, when the values of p and d ($d=2*R_{eff}$) are given, the linear curvature changing rate k is uniquely determined, as
\begin{equation*}
k=\dfrac{\left[2\sqrt{2}\int_0^{\sqrt{\pi p/2}}\sin t^2 dt + \dfrac{2}{\sqrt{\pi p}}\sin{\dfrac{\pi(1-p)}{2}}\right]^2}{d^2}
\end{equation*}
Figure 1(a) illustrates the difference between a full Euler bend (θ=0,p=1), a partial Euler bend (p∈(0,1)), and a circular bend (θ=π,p=0).
In order to plot the bends, 2 Euler's bend and one arc of circonference where considered.
The connection point (the final lenght of the Euler bend) depends on p, such that 
\begin{equation*}
s_F=\dfrac{1}{k R_F}=\sqrt{\dfrac{\pi p}{k}}
\end{equation*}
The centre of the circonference arc also depends on p. Its coordinates are ($x_0,R_{eff}$), where $x_0$ can be obtained through geometric considerations as
\begin{equation*}
x_0=x_F-\dfrac{\sin(p/2)}{\sqrt{k\pi p}}
\end{equation*}
where $x_F$ is the x coordinates of the final point of the Euler bend with lenght $s_F$.
